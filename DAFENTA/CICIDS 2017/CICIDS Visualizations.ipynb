```python
import matplotlib.pyplot as plt
import numpy as np

# Data
models = ['RFC', 'AdaBoost', 'Gradient Boost', 'Extra Trees', 'Logistic Regression', 'KNN']
without_embeddings = [99.85, 98.24, 99.27, 99.84, 91.79, 99.82]
with_embeddings = [99.97, 99.81, 99.87, 99.97, 99.55, 99.98]
accuracy_change = np.subtract(with_embeddings, without_embeddings)  # Calculate change in accuracy

# Plot configuration
fig, ax = plt.subplots(figsize=(12, 8))

# Bar chart
bar_width = 0.4
index = np.arange(len(models))

bars1 = ax.bar(index, without_embeddings, bar_width, label='Without Embeddings', color='skyblue', edgecolor='black')
bars2 = ax.bar(index + bar_width, with_embeddings, bar_width, label='With Embeddings', color='salmon', edgecolor='black')

# Displaying exact accuracy scores and changes on the chart
for i, bar in enumerate(bars1):
    ax.text(bar.get_x() + bar.get_width()/2.0, bar.get_height() + 0.2, f'{without_embeddings[i]:.2f}',
            ha='center', va='bottom', fontsize=10)

for i, bar in enumerate(bars2):
    ax.text(bar.get_x() + bar.get_width()/2.0, bar.get_height() + 0.2, f'{with_embeddings[i]:.2f}',
            ha='center', va='bottom', fontsize=10)

for i, change in enumerate(accuracy_change):
    ax.text(index[i] + bar_width / 2, max(with_embeddings[i], without_embeddings[i]) + 0.5,
            f'{change:.2f}', ha='center', color='green', fontsize=10, fontweight='bold')

# Adjusting Y-axis limits and ticks
ax.set_ylim(90, 101)  # Set the new y-axis limits
ax.set_yticks(np.arange(90, 101, 1))  # Detailed steps from 90 to 100

# Enhancing overall formatting
ax.set_xlabel('Model', fontsize=14)
ax.set_ylabel('Accuracy (%)', fontsize=14)
ax.set_title('Machine Learning Model Accuracies with and without LLM Embeddings', fontsize=16, fontweight='bold')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(models, fontsize=12)
ax.legend(fontsize=12)

# Adding grid for better readability
ax.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

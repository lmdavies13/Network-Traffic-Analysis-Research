```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data based on provided model accuracy information
data = {
    'Model': ['Random Forest', 'AdaBoost', 'Gradient Boost', 'Extra Trees', 'SVC', 'XGBoost'],
    'Original Features Accuracy': [92.78, 92.11, 93.14, 92.56, 92.84, 92.73],
    'LLM Embeddings Accuracy': [93.77, 94.57, 92.14, 93.64, 94.19, 94.05]
}

# Convert data to DataFrame
df = pd.DataFrame(data)

# Create bar plot for model accuracy comparison
fig, ax = plt.subplots(figsize=(10, 6))
bar_width = 0.35
index = range(len(df['Model']))

# Plotting the bars
bars1 = ax.bar(index, df['Original Features Accuracy'], bar_width, label='Original Features Accuracy')
bars2 = ax.bar([i + bar_width for i in index], df['LLM Embeddings Accuracy'], bar_width, label='LLM Embeddings Accuracy')

# Labels and title
ax.set_xlabel('Model', fontsize=12)
ax.set_ylabel('Accuracy (%)', fontsize=12)
ax.set_title('Model Accuracy Comparison with Original Features vs LLM Embeddings', fontsize=14)
ax.set_xticks([i + bar_width / 2 for i in index])
ax.set_xticklabels(df['Model'], rotation=45, ha='right')
ax.legend()

# Display plot
plt.tight_layout()
plt.show()
```

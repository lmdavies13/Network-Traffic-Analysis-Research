```python
import matplotlib.pyplot as plt
import numpy as np

# Data
models = ['RFC', 'AdaBoost', 'Gradient Boost', 'Extra Trees', 'Logistic Regression', 'KNN']
without_embeddings = [96.82, 96.65, 97.37, 96.65, 96.27, 94.33]
with_embeddings = [99.74, 99.03, 99.30, 99.74, 99.27, 99.84]
with_embeddings_pca = [99.41, 98.07, 98.88, 99.11, 97.52, 99.44]
with_encode_embeddings_pca = [99.48, 98.92, 99.0, 99.53, 98.16, 99.88]

# Runtimes
runtimes_without_embeddings = [12, 14, 13, 15, 10, 9]
runtimes_with_embeddings = [18, 17, 19, 16, 20, 21]
runtimes_with_embeddings_pca = [16, 18, 17, 19, 15, 14]
runtimes_with_encode_embeddings_pca = [20, 22, 21, 23, 18, 17]

# Plot configuration
fig, ax = plt.subplots(figsize=(14, 8))

# Bar chart
bar_width = 0.2
index = np.arange(len(models))

bars1 = ax.bar(index, without_embeddings, bar_width, label='Without Embeddings')
bars2 = ax.bar(index + bar_width, with_embeddings, bar_width, label='With Embeddings (Original 384 dimension)')
bars3 = ax.bar(index + 2 * bar_width, with_embeddings_pca, bar_width, label='With Embeddings + PCA (20 dimensions)')
bars4 = ax.bar(index + 3 * bar_width, with_encode_embeddings_pca, bar_width, label='With Embeddings + PCA (20 dimensions)+ Numeric Encode')

# Annotate bars with accuracy scores
for i in range(len(models)):
    ax.text(i, without_embeddings[i] + 0.1, f'{without_embeddings[i]:.2f}', ha='center', va='bottom', fontsize=9)
    ax.text(i + bar_width, with_embeddings[i] + 0.1, f'{with_embeddings[i]:.2f}', ha='center', va='bottom', fontsize=9)
    ax.text(i + 2 * bar_width, with_embeddings_pca[i] + 0.1, f'{with_embeddings_pca[i]:.2f}', ha='center', va='bottom', fontsize=9)
    ax.text(i + 3 * bar_width, with_encode_embeddings_pca[i] + 0.1, f'{with_encode_embeddings_pca[i]:.2f}', ha='center', va='bottom', fontsize=9)

# Line plots
for i, model in enumerate(models):
    x_values = [i, i + bar_width, i + 2 * bar_width, i + 3 * bar_width]
    y_values = [without_embeddings[i], with_embeddings[i], with_embeddings_pca[i], with_encode_embeddings_pca[i]]
    ax.plot(x_values, y_values, marker='o', linestyle='-', label=f'{model} Line')

# Adjusting Y-axis limits and ticks
ax.set_ylim(94, 101)  # Set the new y-axis limits slightly above the max value
ax.set_yticks(np.arange(94, 101, 1))  # Detailed steps from 94.0 to 100.0

# Formatting
ax.set_xlabel('Model')
ax.set_ylabel('Accuracy (%)')
ax.set_title('Machine Learning Model Accuracies with LLM Embeddings (New Dataset with PCA Tests)')
ax.set_xticks(index + 1.5 * bar_width)
ax.set_xticklabels(models)

# Main legend
main_legend = ax.legend(fontsize='small', bbox_to_anchor=(1.05, 1), loc='upper left')

# Runtime legend
runtime_legend_labels = [
    f'Without Embeddings (avg. runtime: {np.mean(runtimes_without_embeddings):.2f}s)',
    f'With Embeddings (Original 384 dimension) (avg. runtime: {np.mean(runtimes_with_embeddings):.2f}s)',
    f'With Embeddings + PCA (20 dimensions) (avg. runtime: {np.mean(runtimes_with_embeddings_pca):.2f}s)',
    f'With Embeddings + PCA (20 dimensions)+ Numeric Encode (avg. runtime: {np.mean(runtimes_with_encode_embeddings_pca):.2f}s)'
]

runtime_patches = [
    plt.Line2D([0], [0], color=bars1[0].get_facecolor(), label=runtime_legend_labels[0]),
    plt.Line2D([0], [0], color=bars2[0].get_facecolor(), label=runtime_legend_labels[1]),
    plt.Line2D([0], [0], color=bars3[0].get_facecolor(), label=runtime_legend_labels[2]),
    plt.Line2D([0], [0], color=bars4[0].get_facecolor(), label=runtime_legend_labels[3])
]

runtime_legend = plt.legend(handles=runtime_patches, fontsize='small', bbox_to_anchor=(0.5, -0.15), loc='upper center', ncol=1)
ax.add_artist(main_legend)
ax.add_artist(runtime_legend)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
datasets = ['DARPA Dataset', 'ISCX Dataset']
average_ml_accuracy = [92.77, 98.93]
average_ml_accuracy_with_llm_embeddings = [93.39, 99.88]
transformer_encoder_classification_model = [93.80, 100]

# Plot configuration
fig, ax = plt.subplots(figsize=(10, 6))

# Bar chart
bar_width = 0.25
index = np.arange(len(datasets))

bars1 = ax.bar(index, average_ml_accuracy, bar_width, label='Best ML Accuracy')
bars2 = ax.bar(index + bar_width, average_ml_accuracy_with_llm_embeddings, bar_width, label='Best ML Accuracy with LLM embeddings')
bars3 = ax.bar(index + 2 * bar_width, transformer_encoder_classification_model, bar_width, label='Best Transformer Encoder Classification Model Accuracy')

# Set y-axis limits
ax.set_ylim(90, 100)

# Add scores above each bar
for i in range(len(datasets)):
    ax.text(index[i], average_ml_accuracy[i] + 0.2, f'{average_ml_accuracy[i]:.2f}', ha='center', va='bottom')
    ax.text(index[i] + bar_width, average_ml_accuracy_with_llm_embeddings[i] + 0.2, f'{average_ml_accuracy_with_llm_embeddings[i]:.2f}', ha='center', va='bottom')
    ax.text(index[i] + 2 * bar_width, transformer_encoder_classification_model[i] + 0.2, f'{transformer_encoder_classification_model[i]:.2f}', ha='center', va='bottom')

# Formatting
ax.set_xlabel('Dataset')
ax.set_ylabel('Accuracy')
ax.set_title('Model Accuracy Comparisons Across Datasets')
ax.set_xticks(index + bar_width)
ax.set_xticklabels(datasets)
ax.legend()

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for accuracy and runtime (in minutes)
datasets = ['Original', 'Embeddings', 'Embeddings + PCA', 'Embeddings + PCA + Numeric', 'Transformer']

# Accuracy for KDD and ISCX
accuracy_kdd = [92.69, 94.06, 93.32, 93.39, 93.80]
accuracy_iscx = [96.35, 99.16, 99.30, 99.48, 100]

# Runtime for KDD and ISCX (in minutes)
runtime_kdd = [23.01, 8*60 + 21 + 41/60, 1*60 + 3 + 50/60, 36.35, 1791/60]
runtime_iscx = [23.49, 13*60 + 27 + 50/60, 1*60 + 12 + 44/60, 46.03, 2914/60]

# Plot configuration
bar_width = 0.35
index = np.arange(len(datasets))

fig, ax1 = plt.subplots(figsize=(12, 8))

# Accuracy bars
bars1_kdd = ax1.bar(index - bar_width/2, accuracy_kdd, bar_width, label='KDD Accuracy', color='b')
bars1_iscx = ax1.bar(index + bar_width/2, accuracy_iscx, bar_width, label='ISCX Accuracy', color='g')

# Create a second y-axis for runtime
ax2 = ax1.twinx()

# Runtime bars
bars2_kdd = ax2.bar(index - bar_width/2, runtime_kdd, bar_width, label='KDD Runtime', color='b', alpha=0.4, hatch='//')
bars2_iscx = ax2.bar(index + bar_width/2, runtime_iscx, bar_width, label='ISCX Runtime', color='g', alpha=0.4, hatch='//')

# Accuracy axis settings
ax1.set_xlabel('Dataset and Method')
ax1.set_ylabel('Accuracy (%)')
ax1.set_ylim([90, 105])  # Set y-axis for accuracy
ax1.set_title('Accuracy and Runtime Comparison Across Datasets and Methods')

# Runtime axis settings
ax2.set_ylabel('Runtime (Minutes)')
ax2.set_ylim([0, max(max(runtime_kdd), max(runtime_iscx)) * 1.1])

# X-axis labels
ax1.set_xticks(index)
ax1.set_xticklabels(datasets)

# Add legends
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data: Accuracy values for the best model per method (example values)
datasets = ['DARPA Dataset', 'ISCX Dataset']
ml_accuracy = [92.77, 98.93]  # Best ML accuracy
ml_accuracy_with_embeddings = [93.39, 99.88]  # Best ML accuracy with embeddings
transformer_accuracy = [93.80, 100]  # Transformer accuracy

# Names of the best-performing models for annotations (example values)
best_ml_model = ['Gradient Boost', 'Gradient Boost']
best_ml_model_with_embeddings = ['AdaBoost', 'KNN']

# Plot configuration
fig, ax = plt.subplots(figsize=(10, 6))

# Bar chart
bar_width = 0.25
index = np.arange(len(datasets))

# Plot each set of bars for ML and transformer models
bars1 = ax.bar(index, ml_accuracy, bar_width, label='ML Accuracy')
bars2 = ax.bar(index + bar_width, ml_accuracy_with_embeddings, bar_width, label='ML Accuracy with LLM embeddings')
bars3 = ax.bar(index + 2 * bar_width, transformer_accuracy, bar_width, label='Transformer Encoder Accuracy')

# Set y-axis limits with extra space above 100
ax.set_ylim(90, 101.5)

# Add accuracy scores above each bar and model annotations
for i in range(len(datasets)):
    ax.text(index[i], ml_accuracy[i] + 0.2, f'{ml_accuracy[i]:.2f}', ha='center', va='bottom')
    ax.text(index[i] + bar_width, ml_accuracy_with_embeddings[i] + 0.2, f'{ml_accuracy_with_embeddings[i]:.2f}', ha='center', va='bottom')
    ax.text(index[i] + 2 * bar_width, transformer_accuracy[i] + 0.2, f'{transformer_accuracy[i]:.2f}', ha='center', va='bottom')

    # Annotate which model achieved the best ML results
    ax.text(index[i], ml_accuracy[i] - 0.6, f'{best_ml_model[i]}', ha='center', fontsize=10, color='blue')
    ax.text(index[i] + bar_width, ml_accuracy_with_embeddings[i] - 0.6, f'{best_ml_model_with_embeddings[i]}', ha='center', fontsize=10, color='green')

# Formatting
ax.set_xlabel('Dataset')
ax.set_ylabel('Accuracy')
ax.set_title('Model Accuracy Comparisons Across Datasets')
ax.set_xticks(index + bar_width)
ax.set_xticklabels(datasets)
ax.legend()

# Display the plot
plt.show()
```
